Livro:The Prince, by Nicolo Machiavelli

```
import requests
from bs4 import BeautifulSoup
import nltk
```

```
r = requests.get("https://www.gutenberg.org/cache/epub/1232/pg1232-images.html")

r.encoding = 'utf-8'

html = r.text

print(html[0:2000])
```

```
soup = BeautifulSoup(html, 'html.parser')

text = soup.get_text()

print(text[32000:34000])
```

```
tokenizer = nltk.tokenize.RegexpTokenizer('\w+')

tokens = tokenizer.tokenize(text)

print(tokens[0:8])
```

```
words = []

for word in tokens:
  words.append(word.lower())

print(words[0:8])
```

```
from nltk.corpus import stopwords
nltk.download('stopwords')
sw = stopwords.words('english')
```

```
words_ns =  [word for word in words if word not in sw]
```

```
print(words_ns[0:8])
```

```
freqdist = nltk.FreqDist(words_ns)
```

```
freqdist
```

```
%matplotlib inline

freqdist.plot(500)
```

Livro: Les Misérables by Victor Hugo

```
import requests
from bs4 import BeautifulSoup
import nltk
```

```
r = requests.get("https://www.gutenberg.org/cache/epub/135/pg135-images.html")

r.encoding = 'utf-8'

html = r.text

print(html[0:2000])
```

```
soup = BeautifulSoup(html, 'html.parser')

text = soup.get_text()

print(text[32000:34000])
```

```
tokenizer = nltk.tokenize.RegexpTokenizer('\w+')

tokens = tokenizer.tokenize(text)

bigrams = list(nltk.bigrams(tokens))

print(bigrams[:8])
```

```
words = [word.lower() for word in tokens]

bigrams = list(nltk.bigrams(words))

print(bigrams[:8])
```

```
from nltk.corpus import stopwords
nltk.download('stopwords')
sw = stopwords.words('english')
```

```
filtered_bigrams = [bigram for bigram in bigrams if bigram[0] not in sw and bigram[1] not in sw]
```

```
print(filtered_bigrams)
```

```
freqdist = nltk.FreqDist(filtered_bigrams)
```

```
freqdist
```

```
%matplotlib inline

freqdist.plot(500)
```

# As palavras mais frequentes de alguma forma estão relacionados com o enredo do livro?
  Sim. As palavras mais frequentes estão relacionadas ao enredo de "Os Miseráveis" de Victor Hugo, incluindo o protagonista Jean Valjean, eventos importantes e referências ao projeto Gutenberg.
  A presença frequente de bigramas como ('jean', 'valjean'), que se refere ao protagonista da história, é uma indicação clara de que o enredo gira em torno desse personagem central. Além disso, palavras como 'rue' (rua), 'monsieur' (senhor), 'old' (velho), 'la' (a), 'chapter' (capítulo) e 'gutenberg' indicam elementos importantes do enredo, como cenários, personagens, estrutura da narrativa e referências à publicação do livro no projeto Gutenberg.
